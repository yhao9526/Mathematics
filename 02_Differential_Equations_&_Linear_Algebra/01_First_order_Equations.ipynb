# 1 一阶方程

## 1.1 一阶线性微分方程

### ODE标准式

$$
\begin{cases}
\displaystyle \frac{dy}{dt} + g(t)y = f(t) \\[10pt]
y(0) = y_0
\end{cases}
$$
求解：利用积分因子 $u(t)=e^{\int_0^t g(s)ds}$，我们可以将原方程转化为全微分形式并进行积分：

$$
\begin{aligned}
&\frac{d}{dt} \left[ y(t) e^{\int_0^t g(s)ds} \right] = f(t) e^{\int_0^t g(s)ds} \\[10pt]
&y(t) e^{\int_0^t g(s)ds} = \int_0^t f(s) e^{\int_0^s g(\tau)d\tau} ds + y(0) \\[10pt]
&y(t) = e^{-\int_0^t g(s)ds} \left[ \int_0^t f(s) e^{\int_0^s g(\tau)d\tau} ds \right] + y(0)e^{-\int_0^t g(s)ds}
\end{aligned}
$$

> **注释：**
> * **第一步**：两边同时乘以积分因子，将左边凑成乘法法则的逆运算。
> * **第二步**：对两边关于 $t$ 从 $0$ 到 $t$ 积分，注意右侧积分变量改为 $s$ 以示区分。
> * **第三步**：两边同乘 $e^{-\int_0^t g(s)ds}$，解出 $y(t)$。

## 1.2 一阶代换方程

### 伯努利ODE

$$
\begin{cases}
y' = P(x)y + Q(x)y^n \\[10pt]
n \neq 0, 1
\end{cases}
$$
求解：通过对原方程两边同除以 $y^n$，并引入新变量 $v = y^{1-n}$ 进行代换：
$$
\begin{aligned}
&\frac{y'}{y^n} = P(x) \frac{1}{y^{n-1}} + Q(x) \\[10pt]
&\text{令 } v(x) = y^{1-n}(x), \text{ 则通过链式法则有：} \\[5pt]
&v' = (1-n) y^{-n} \cdot y' \implies \frac{y'}{y^n} = \frac{v'}{1-n} \\[10pt]
&\text{代入上式得：} \\[5pt]
&\frac{v'}{1-n} = P(x)v + Q(x) \\[10pt]
&v' + (n-1)P(x)v = -(n-1)Q(x)
\end{aligned}
$$
化为标准一阶线性方程；
> **注释：** 建模非线性约束

### 一阶齐次ODE

$$
y' = F\left(\frac{y}{x}\right)
$$

求解：变量替换—我们引入新变量 $u$，将 $y$ 表示为 $x$ 的缩放：  
$$
\begin{aligned}
&\text{令 } u = \frac{y}{x} \implies y = ux \\[10pt]
&\text{对 } x \text{ 求导（利用乘法法则）：} y' = u + x \frac{du}{dx} \\[10pt]
&\text{代入原方程：} u + x \frac{du}{dx} = F(u) \\[10pt]
&\text{整理得可分离变量形式：} \frac{du}{F(u) - u} = \frac{dx}{x}
\end{aligned}
$$
此时方程两边已完全分离，只需两边同时积分即可求得 $u(x)$，最后代回 $y = ux$ 即可。
> **注释**：建模角度变化

### Example

设原点 $O$ 处有一灯光照射，一艘船在航行过程中，其**航迹切线**与**灯光照射方向**（极径方向）的夹角恒为 $\alpha = \frac{\pi}{4}$。试求船的航迹方程。

**对船：**
如上图，在极坐标系 $(r, \theta)$ 中，利用几何关系 $\tan \alpha = r \frac{d\theta}{dr}$，已知 $\alpha = \frac{\pi}{4}$，则：

$$
\begin{aligned}
&r \frac{d\theta}{dr} = \tan\left(\frac{\pi}{4}\right) = 1 \\[10pt]
&\frac{dr}{r} = d\theta
\end{aligned}
$$

**求解：**
$$
\begin{aligned}
&\int \frac{1}{r} dr = \int d\theta \\[10pt]
&\ln |r| = \theta + C \\[10pt]
&r = Ce^{\theta}
\end{aligned}
$$

**结论：** 船的航迹为等角螺线—对数螺线。

## 1.3 一阶自治方程

### 自治ODE

系统变化不依赖于时间 $t$，即变化率仅由系统当前状态 $y$ 决定。

$$
\frac{dy}{dt} = f(y)
$$
**求解：**
获取方程关键信息，整合求解。在很多情况下，自治方程往往求不出解析解，但通过分析其定性特征，可以有效地**指导决策**。
$$
\begin{aligned}
&\text{令 } f(y_0) = 0 \\[10pt]
&\text{若存在常数 } y_0 \text{ 满足上式，则：} \\[10pt]
&y(t) \equiv y_0 \text{ 是方程的解（称为平衡解或临界解）。}
\end{aligned}
$$

通过寻找这些平衡点，我们可以分析系统在不同状态下的稳定性（如趋向平衡或远离平衡），而无需知道具体的 $y(t)$ 表达式。
> **注释：** 建模变化自相关

### Example

$$ y=养殖鱼主总数，k=增长率，k=k(y)=a-by，h=捕捞量(常数) $$

**求解：** 根据系统变化自相关的特性，建立如下方程：

$$
\frac{dy}{dt} = ay - by^2 - h
$$
如下图，该方程描述了鱼群在受限环境下的增长与固定强度捕捞之间的关系。
